```python
X = df.drop(['diagnosis'], axis=1)
y = df['diagnosis']
```

## Normalize the Data

```python
from sklearn.preprocessing import StandardScaler

# Create a Object of StandardScaler
SS = StandardScaler()

# Fit the Value of Transform the Scaling 
X_scaled = SS.fit_transform(X)
```

```python
X_scaled
```

## Split the Data

```python
from sklearn.model_selection import train_test_split

# Split the data to train and test

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.30, random_state=0)
```

```python
type(X_train)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

# Create a Logistic Model
model = LogisticRegression()

# Fit the train of model
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)
y_predict
```

```python
model_score = model.score(X_test, y_test)
model_score
```

```python
# Predict probabilities on the validation set
y_probs = model.predict_proba(X_test)[:, 1] # Find probability of class 1
y_probs
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_predict)
print(cm)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_predict)
print(f' Accuracy : {accuracy}')
```

```python
# Assuming you have the confusion matrix stored in the variables tp, tn, fp, fn
tp = 60
tn = 107
fp = 1
fn = 3

# Calculate accuracy
accuracy = (tp + tn) / (tp + tn + fp + fn)
print("Accuracy:", accuracy)

```

```python
# Calculate precision

precision = tp / (tp + fp)
print("Precision:", precision)
```

```python
# Calculate recall (sensitivity)

recall = tp / (tp + fn)
print("Recall:", recall)
```

```python
# Calculate specificity

specificity = tn / (tn + fp)
print("Specificity:", specificity)
```

```python
# Calculate F1 score

f1_score = 2 * (precision * recall) / (precision + recall)
print("F1 Score:", f1_score)
```

# AUC - ROC Curve Score

```python
# Step 5: Calculate AUC-ROC score:
from sklearn.metrics import roc_auc_score,roc_curve

auc_roc = roc_auc_score(y_test, y_predict)
print(f"AUC-ROC Score: {auc_roc}")
```

```python
# Step 6: Plot ROC curve
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve


fpr, tpr, _= roc_curve(y_test, y_predict)

# Calculates the False Positive Rate (FPR) and True Positive Rate (TPR) 
#at various threshold settings for the given y_test (actual labels) 
#and y_predict (predicted probabilities or scores).

plt.figure()
#Creates a new figure for the plot.

plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % auc_roc)
#Plots the ROC curve using the FPR (x-axis) and TPR (y-axis) values.
#color='darkorange' sets the color of the curve to dark orange.
# lw=2 sets the line width to 2.
# label='ROC curve (area = %0.2f)' % auc_roc adds a label to the curve showing the AUC-ROC score.

plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
# Plots a diagonal line from (0,0) to (1,1) representing the ROC curve of a purely random classifier.

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
# Sets the limits for the x-axis and y-axis respectively.
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
y_probs = model.predict_proba(X_test)[:, 1]

# Define a function to tune the threshold
def threshold_tuner(y_true, y_prob, threshold):
    return np.where(y_prob > threshold, 1, 0)

# Define a range of threshold values to try
thresholds = np.arange(0, 1, 0.05)

# Iterate over the threshold values and evaluate the model
best_threshold = 0
best_score = 0
for threshold in thresholds:
    y_pred = threshold_tuner(y_test, y_probs, threshold)
    score = accuracy_score(y_test, y_pred)
    if score > best_score:
        best_score = score
        best_threshold = threshold

# Print the best threshold value
print("Best Threshold:", best_threshold)
```
